### Pandas Series



*   Create series type of data using various data structures collection - array, list, tuples, dict etc.


```python
# 1D representation of data - Series
# Series does not provide many inbuilt functions.
data = pd.Series([10, 20, 30, 40, 50, 60, 70])
print (data)
```

```python
data.index
```

```python
data.values
```

```python
# Collection Of Series
data_series = {'Column1' : pd.Series([100, 200, 300, 400, 500, 600, 700]),
               'Column2' : pd.Series([10, 20, 30, 40, 50, 60, 70])
}
print(data_series)
```

```python
# The above is bad data representation.
# Typecast series to DataFrames.
pd.DataFrame(data_series)
```
